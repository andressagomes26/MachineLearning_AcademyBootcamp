```python
imgOriginal = cv2.imread('shine54.jpg', 1)
imgSuavizada = cv2.GaussianBlur(imgOriginal, (13,13), 3)
imgDetalhes = 3 * cv2.subtract(imgOriginal, imgSuavizada)
imgRealcada = cv2.add(imgOriginal, imgDetalhes)

imgOriginal = imgOriginal[::2,::2] # Diminui a imagem 
imgSuavizada = imgSuavizada[::2,::2] # Diminui a imagem 
imgDetalhes = imgDetalhes[::2,::2] # Diminui a imagem 
imgRealcada = imgRealcada[::2,::2] # Diminui a imagem

cv2.imshow("Original", imgOriginal)
cv2.imshow("Suavizada", imgSuavizada)
#cv2.imshow("Bordas", imgDetalhes)
cv2.imshow("Realcada", imgRealcada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
